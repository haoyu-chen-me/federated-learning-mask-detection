```python
import pandas as pd
import matplotlib.pyplot as plt

# 加载 CSV 文件
df = pd.read_csv("accuracy_log_CIFAR.csv", names=["round", "accuracy"])

# 画图
plt.figure(figsize=(8, 5))
plt.plot(df["round"], df["accuracy"], marker='o', label="Global Accuracy")
plt.title("Federated Learning on CIFAR-10")
plt.xlabel("Communication Round")
plt.ylabel("Accuracy")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
